```python
k=0.1
a=-5.2
pp=log(a+5.5)
c=3
t=log(c+5.5)
p=plot(log(x+5.5), (x, -5.4, 5), zorder=3, thickness=4, color=oacmaroon, axes=False)
l=line([(-5.4, 0), (5, 0)], color="black", thickness=2)+line([(0,log(0.1)),(0,log(10.5))], color="black", thickness=2) # axes
tx=line([(a, -k), (a, k)], thickness=2, color="black") + line([(-4.5, -k), (-4.5, k)], thickness=2, color="black")+line([(c, -k), (c, k)], thickness=2, color="black")# ticks along the x-axis
ty=line([(-k, pp), (k, pp)], thickness=2, color="black") + line([(-k, log(5.5)), (k, log(5.5))], thickness=2, color="black") + line([(-k, t), (k, t)], thickness=2, color="black") # ticks along the y-axis
lx=text("$a$", (a, -k-0.1), fontsize=20, color="black") + text("$b$", (-4.5, -k-0.1), fontsize=20, color="black") +text("$c$", (c, -k-0.1), fontsize=20, color="black")+text("$x$", (5.2, 0), fontsize=20, color="black")
ly=text("$p$", (-k-0.1, pp), fontsize=20, color="black") + text("$r$", (-k-0.1, log(5.5)), fontsize=20, color="black") + text("$t$", (-k-0.1, t), fontsize=20, color="black")+text("$y$", (0, log(10.5)+0.2), fontsize=20, color="black") 
ll=text("$K$", (a+0.3, log(a+5.5)), fontsize=20, color="black") + text("$L$", (c, t+0.2), fontsize=20, color="black")
pt=point((a, log(a+5.5)), size=120, color="black", zorder=4) + point((c, t), size=120, color="black", zorder=4) 
p2=p+l+tx+ty+lx+ly+pt+ll
show(p2)
p2.save('prob21.png')
```
